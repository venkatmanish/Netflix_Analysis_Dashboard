```python
from sklearn.impute import KNNImputer

# Handles missing values for categorical/text columns
data['Director'] = data['Director'].fillna('Unknown')
data['Cast'] = data['Cast'].fillna('No Cast Information')
data['Country'] = data['Country'].fillna('Unknown')
data['Rating'] = data['Rating'].fillna('Unknown')

# Converts Release_Date to datetime and handle missing dates
data['Release_Date'] = pd.to_datetime(data['Release_Date'], errors='coerce')
data['Release_Date'] = data['Release_Date'].fillna(data['Release_Date'].mode()[0])

# Preprocess Duration into numerical format
def standardize_duration(duration):
    if 'min' in duration:
        return int(duration.replace(' min', ''))
    elif 'Season' in duration:
        return int(duration.split()[0]) * 60  # Approximate 1 season = 60 mins
    return np.nan

data['Duration'] = data['Duration'].apply(standardize_duration)

# KNN Imputer to numerical columns
numerical_cols = data.select_dtypes(include=[np.number])
if numerical_cols.isnull().sum().sum() > 0:  # Check if any numerical column has missing values
    knn_imputer = KNNImputer(n_neighbors=5)
    numerical_cols_imputed = knn_imputer.fit_transform(numerical_cols)
    data[numerical_cols.columns] = numerical_cols_imputed

# Handles any remaining missing values in Duration using median
data['Duration'] = data['Duration'].fillna(data['Duration'].median())
```

```python
print(data.dtypes)
```

```python
# Handle Outliers in Duration
Q1 = data['Duration'].quantile(0.25)
Q3 = data['Duration'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and replace outliers with NaN
outliers = data[(data['Duration'] < lower_bound) | (data['Duration'] > upper_bound)]
data['Duration'] = data['Duration'].where((data['Duration'] >= lower_bound) & (data['Duration'] <= upper_bound), np.nan)

print(f"Number of outliers detected and removed: {len(outliers)}")

# Impute Missing Values in Duration
# Use median or KNN imputation to re-fill missing values
if data['Duration'].isnull().sum() > 0:
    data['Duration'] = data['Duration'].fillna(data['Duration'].median())

# Duration_per_Episode
# Ensure derived column is recalculated after handling Duration
data['Duration_per_Episode'] = data['Duration'] / data['Type'].apply(lambda x: 1 if x == 0 else 10)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['Category'] = label_encoder.fit_transform(data['Category'])
data['Type'] = label_encoder.fit_transform(data['Type'])
```

```python
# Extract year, month, and day from Release_Date
data['Release_Year'] = data['Release_Date'].dt.year
data['Release_Month'] = data['Release_Date'].dt.month
data['Release_Day'] = data['Release_Date'].dt.day

print(data.head())
```

**SQL Database Design and Querying**

```python
import sqlite3
import pandas as pd
```

```python
# 1NF: Ensuring atomicity
# Multi-valued columns 'Cast' and 'Type' are split into atomic values 
# and stored in separate supporting tables (cast_data and type_data).

# Normalize 'Cast' into a supporting table
cast_data = data[['Show_Id', 'Cast']].dropna()
cast_data = cast_data.assign(Cast_Member=cast_data['Cast'].str.split(', ')).explode('Cast_Member')
cast_data.drop(columns=['Cast'], inplace=True)

# Normalize 'Type' into a supporting table
type_data = data[['Show_Id', 'Type']].dropna()

# Ensure 'Type' contains only string values
type_data['Type'] = type_data['Type'].astype(str).fillna('')

# Split and explode 'Type' into atomic values
type_data = type_data.assign(Type=type_data['Type'].str.split(', ')).explode('Type')


# Main table retains atomic attributes and does not include multi-valued columns
# 3NF: Derived columns 'Release_Year', 'Release_Month', 'Release_Day' are removed
# because they can be calculated dynamically from 'Release_Date'.
main_data = data.drop(columns=['Cast', 'Type', 'Release_Year', 'Release_Month', 'Release_Day'])
```

```python
# Create the main table for normalized data
# 1NF: Atomic values are stored in each column.
# 2NF: Every attribute is fully dependent on the primary key (Show_Id).
# 3NF: Derived attributes like 'Release_Year', 'Release_Month', 'Release_Day' are not stored 
# and can be dynamically calculated from 'Release_Date'.

cursor = conn.cursor()
create_main_table_query = """
CREATE TABLE IF NOT EXISTS netflix_data (
    Show_Id TEXT PRIMARY KEY,
    Category TEXT,
    Title TEXT,
    Director TEXT,
    Country TEXT,
    Release_Date DATE,
    Rating TEXT,
    Duration INTEGER,
    Description TEXT
);
"""
cursor.execute(create_main_table_query)
conn.commit()
print("Main table created successfully!")
```

```python
# Create supporting table for Cast
# 1NF: Each cast member is stored in a separate row in cast_data.
# 2NF: Each row is fully dependent on the primary key (Show_Id).
create_cast_table_query = """
CREATE TABLE IF NOT EXISTS cast_data (
    Show_Id TEXT,
    Cast_Member TEXT,
    FOREIGN KEY (Show_Id) REFERENCES netflix_data(Show_Id)
);
"""
cursor.execute(create_cast_table_query)

# Create supporting table for Type
# 1NF: Each type is stored in a separate row in type_data.
# 2NF: Each row is fully dependent on the primary key (Show_Id).
create_type_table_query = """
CREATE TABLE IF NOT EXISTS type_data (
    Show_Id TEXT,
    Type TEXT,
    FOREIGN KEY (Show_Id) REFERENCES netflix_data(Show_Id)
);
"""
cursor.execute(create_type_table_query)
conn.commit()
print("Supporting tables (cast_data and type_data) created successfully!")
```

```python
# Insert main table data into the database
# 1NF and 2NF are satisfied since the data is atomic and attributes depend on Show_Id.
main_data.to_sql('netflix_data', conn, if_exists='replace', index=False)
```

**Advanced Data Analysis with Machine Learning**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Connect to SQLite database
conn = sqlite3.connect('netflix_data_normalized.db')

# Retrieve main dataset
query = "SELECT * FROM netflix_data;"
data = pd.read_sql(query, conn)

# Preview the data
print("Main Data Preview:")
print(data.head())
```

```python
# Encode categorical features for modeling
data['Category'] = data['Category'].map({'Movie': 0, 'TV Show': 1})
data['Rating'] = data['Rating'].astype('category').cat.codes
data['Country'] = data['Country'].astype('category').cat.codes
data['Description_Length'] = data['Description'].str.len()  # Feature: Length of description
```

```python
# Select features and target
X = data[['Rating', 'Country', 'Duration', 'Description_Length']]
y_classification = data['Category']  # For classification task
y_regression = data['Duration']      # For regression task
```

```python
# Predicting whether the content is a Movie or TV Show
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_classification, test_size=0.2, random_state=42)

# Train a Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Classification Accuracy: {accuracy:.2f}")
```

```python
# Predicting the duration of content
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_regression, test_size=0.2, random_state=42)

# Train a Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict and evaluate
y_pred = lr_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Regression Mean Squared Error: {mse:.2f}")
print(f"Regression R-squared: {r2:.2f}")
```

```python
# Grouping similar content based on feature
# Normalize numerical features for clustering
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_normalized)

# Add cluster labels to the dataset
data['Cluster'] = clusters

# Visualize the clusters
sns.scatterplot(data=data, x='Duration', y='Description_Length', hue='Cluster', palette='Set2')
plt.title('Clustering of Content')
plt.xlabel('Duration')
plt.ylabel('Description Length')
plt.show()
```

```python
#In summary
print("""
Summary of Machine Learning Results:
1. Classification: Random Forest achieved an accuracy of {:.2f}, showing good prediction capability for content type (Movie or TV Show).
2. Regression: Linear Regression achieved an R-squared of {:.2f}, indicating a moderate ability to predict content duration based on features.
3. Clustering: K-Means identified 3 distinct clusters based on duration and description length, highlighting groups of similar content.
""".format(accuracy, r2))
```
